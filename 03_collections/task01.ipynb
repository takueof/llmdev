```python
grades = [85, 90, 78, 92, 88]
ngrades = len(grades)
total = sum(grades)
average = float(total) / float(ngrades)
max_grade = max(grades)
min_grade = min(grades)
above_average_grades = [grade for grade in grades if average <= float(grade)]
nabove_average_grades = len(above_average_grades)

print(f"平均点：{average}\n",
      f"最高点：{max_grade}\n",
      f"最低点：{min_grade}\n",
      f"平均以上の学生数：{nabove_average_grades}")
```
