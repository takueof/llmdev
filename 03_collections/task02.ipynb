```python
words = list()
alphabets = dict.fromkeys("abcdefghijklmnopqrstuvwxyz", 0)

while True:
    word = input("英単語を入力してください：").strip()
    if word == "":
        break
    words.append(word)
    chars = word[:]
    for char in chars:
        if char in alphabets:
            alphabets[char] += 1

sorted_words = sorted(words)

print(f"入力した英単語：{sorted_words}")

keys = alphabets.keys()
for key in keys:
    value = alphabets[key]
    if 0 < value:
        print(f"{key}が{value}個ありました")
```
