```python
def validate_user_name(name = ""):
    name_length = len(name);
    if (name_length < 5 or 15 < name_length):
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください", flush=True)
        return False
    elif " " in name:
        print("エラー: ユーザー名に空白を含めることはできません", flush=True)
        return False
    elif not name.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます", flush=True)
        return False
    return True

print("ユーザー名を入力してください:", end="", flush=True)
user_name = input("")
valid = validate_user_name(user_name)

if valid:
    print("ユーザー名は有効です", flush=True)
else:
    print("不正なユーザー名です", flush=True)
```
